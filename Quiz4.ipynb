## Import Data

<font size='4'>

* Three lists named `study_id_ls`, `measure_date_ls`, and `duration_ls` have been initialized.
* Write a for loop to extract and append the values of `study_id`, `measure_date`, and `duration` to `study_id_ls`, `measure_date_ls`, and `duration_ls`, respectively.

```python
# empty lists? i think

study_id_ls = []
measure_date_ls = []
duration_ls = []

# put all the data files into one list

all_data = [data_1, data_2, data_3, data_4, data_5, data_6]

# writing for loop for extraction and appending values

for one_file in all_data:
    # find the value called "study_id" & add to the list
    study_id_ls.append(one_file['study_id'][0])
    
    # find the value called "measure_date" & add to the list
    measure_date_ls.append(one_file['measure_date'][0])
    
     # find the value called "duration" & add to the list
    duration_ls.append(one_file['duration'][0])
```

<font size='4'>

* Create a new Pandas Dataframe combining three lists and name it `input_df`.

```python
# create dataframe that combines the lists

input_df = pd.DataFrame({
    'study_id': study_id_ls,
    'measure_date': measure_date_ls,
    'duration': duration_ls
})
```

## Data Manipulation

<font size='4'>

* Reorder the dataframe by `measure_date` (earliest to latest) and then by `duration` (shortest to longest).

```python
# have to change measure_date from text to dates to complte this

input_df['measure_date'] = pd.to_datetime(input_df['measure_date'])

# sort data by date then by duration

input_df = input_df.sort_values(by=['measure_date', 'duration'])
```

<font size='4'>

* Create a new column called `relative_time` by calculating the time difference in days between each of the measure date and the earliest one.

```python
# find earliest date

earliest_date = input_df['measure_date'].min()

# calculate how many days each date is from earliest date

input_df['relative_time'] = (input_df['measure_date'] - earliest_date).dt.days
```

<font size='4'>

* Reset the index column and remove the old index.

```python
# reset row numbers to start from 0 so that it can be read normally; this should rmeove old index

input_df = input_df.reset_index(drop=True)
```
