```python
# import dataset

df = pd.read_csv("/Users/amyhunt/PycharmProjects/BIOS584/Students Social Media Addiction.csv")

# calculate general summary stats

summary_stats = df.describe()
print(summary_stats)

categorical_summary = df.describe(include='object')
print("Summary using describe(include='object'):")
print(categorical_summary)
```

```python
# summary stats for usage hrs

column_name = 'Avg_Daily_Usage_Hours'

# mean of usage hrs
mean_column = df[column_name].mean()
print(f"Mean of Usage Hrs/Day: {mean_column}")

# median of usage hrs
median_column = df[column_name].median()
print(f"Median of Usage Hrs/Day: {median_column}")

# mode of usage hrs
mode_column = df[column_name].mode()
print(f"Mode of Usage Hrs/Day: {mode_column}")

# range of usage hrs
data_range = df[column_name].max() - df[column_name].min()
print(f"Range of Usage Hrs/Day: {data_range}")

# variance of usage hrs
variance = df[column_name].var()
print(f"Variance of Usage Hrs/Day: {variance}")

# std of usage hrs
std_dev = df[column_name].std()
print(f"Standard Deviation of Usage Hrs/Day: {std_dev}")

# IQR of usage hrs
Q1_usage = df[column_name].quantile(0.25)
Q3_usage = df[column_name].quantile(0.75)
iqr = Q3_usage - Q1_usage
print(f"IQR of Usage Hrs/Day: {iqr}")
```

```python
# summary stats for sleep per night

column_name2 = 'Sleep_Hours_Per_Night'

# mean of sleep
mean_column2 = df[column_name2].mean()
print(f"Mean of Sleep/Night: {mean_column2}")

# median of sleep
median_column2 = df[column_name2].median()
print(f"Median of Sleep/Night: {median_column2}")

# mode of sleep
mode_column2 = df[column_name2].mode()
print(f"Mode of Sleep/Night: {mode_column2}")

# range of sleep
data_range2 = df[column_name2].max() - df[column_name2].min()
print(f"Range of Sleep/Night: {data_range2}")

# variance of sleep
variance2 = df[column_name2].var()
print(f"Variance of Sleep/Night: {variance2}")

# std of sleep
std_dev2 = df[column_name2].std()
print(f"Standard Deviation of Sleep/Night: {std_dev2}")

# IQR of sleep
Q1_sleep = df[column_name2].quantile(0.25)
Q3_sleep = df[column_name2].quantile(0.75)
iqr2 = Q3_sleep - Q1_sleep
print(f"IQR of Sleep/Night: {iqr2}")
```

## Objective Three: Statistical Analysis Plan (SAP)

##### To address the first aim of this analysis, which seeks to examine whether average daily social media usage (measured in hours) is associated with the amount of sleep an individual receives per night, a simple linear regression model will be utilized. This model is appropriate because both variables—average daily usage and sleep hours—are continuous, and it will provide a clear indication of how sleep duration changes with each progressive hour of increased social media use. The second aim is to assess whether average daily social media usage is associated with the number of interpersonal conflicts individuals report as being related to social media. For this aim, a simple linear regression model will again be used, as both variables can be treated as continuous; while the conflict variable is technically a count, it is still acceptable for simple linear regression and will allow for evaluating how conflict frequency changes with each additional hour of usage. The third aim seeks to examine whether average daily social media usage is associated with relationship status, categorized as single, complicated, or in a relationship. Because the dependent variable is categorical with three unordered categories, a multinomial logistic regression model will be used to evaluate whether daily usage predicts relationship status.
##### Before these models are fit, the analysis will begin with a general review of the dataset to assess accuracy and completeness. This process will include checking for missing entries, verifying that all variables contain plausible values, and identifying any potential outliers or skews. Descriptive statistics (such as means, medians, standard deviations) will be calculated to summarize each variable, and data visualizations such as histograms and boxplots will be used to identify irregularities in the data. Correlation analyses will also be conducted to explore relationships among continuous variables, providing an initial understanding of the dataset and guiding decisions for subsequent modeling.
##### All data cleaning, descriptive analyses, and modeling will be conducted in Python. The “pandas” and “numpy” packages will be used for data management and numerical operations, while “matplotlib” will be used to generate visualizations. Inferential testing will rely on the “scipy.stats” package for functions such as correlation coefficients and normality checks. Throughout the analysis process, functions including describe(), scatterplot(), boxplot(), and pearsonr() will likely be used.
##### Univariate analyses will be conducted for all variables prior to modeling. For continuous variables (such as daily usage, sleep hours, and conflicts) descriptive statistics (mean, standard deviation, median, range) will be computed, and relevant assumptions will be checked. This includes assessing linearity, homoscedasticity, and normality, along with using histograms to review distributional patterns. For the categorical variable relationship status, frequency tables will be generated to summarize category counts, and a bar chart will be used to visualize the distribution. A chi-square goodness-of-fit test may also be used to assess whether observed category frequencies differ from expected patterns.
##### To effectively present the results, both tables and visualizations will be included. Summary tables will report descriptive statistics for all key variables: average daily usage hours, sleep duration, conflicts, and relationship status. Scatterplots will display relationships between social media usage and sleep, as well as between daily usage and reported conflicts. Boxplots will illustrate how conflict frequency varies across categories of social media usage. In addition, a correlation assessment will be conducted to visualize the strength and direction of relationships among continuous variables. These tables and figures will collectively help identify trends and patterns in the data and support interpretation of the statistical models.
